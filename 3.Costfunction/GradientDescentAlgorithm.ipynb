## Prediction function using Gradient descent algorithm and Mean squared error

```python
import pandas as pd
```

```python
df=pd.read_csv('test_scores.csv')
df.head()
```

```python
x=df.math
y=df.cs
```

```python
# math scores and cs scores relates somehow, so find the correlation between them
# in order to find correlation, calculate coefficient and intercept
# y = mx + c

n=len(df)
iters=1000
learning_rate=0.0001

m=c=0
for i in range(iters):
    
    y_pred= m*x + c
    
    cost=(1/n)*sum([val**2 for val in (y-y_pred)])
    mtemp=-(2/n)*sum(x*(y-y_pred))
    ctemp=-(2/n)*sum(y-y_pred)
    
    m = m - learning_rate*mtemp
    c = c - learning_rate*ctemp
    
    print("iter {}: m {}, c {}, cost {}".format(i,m,c,cost))
```

```python
print("m={}, c={}".format(m,c))
```
